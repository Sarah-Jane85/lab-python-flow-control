```python
# define products
products = ["t-shirt", "mug", "hat", "book", "keychain"]
# create empty dictionary
inventory = {}

inventory = {product: int(input(f"Please input the amount of the product {product}")) for product in products}
print(inventory)
```

```python
#make emty set
customer_orders = set()

#keep asking until no
while True:
    product = input("which product would you like to order?")
    #add to customer_orders
    customer_orders.add(product)
    #ask something else?
    next_order = input("would you like to order something else? Please answer 'yes' or 'no':").strip().lower() 

    if next_order == 'no':
        break

#print order
print("Total order:")
for order in customer_orders:
    print(order)
    
```

```python
#establish total number of orders
tot_products_ordered = len(customer_orders)
#establish total number of products
tot_products = sum(inventory.values())
#calculate percentage
percentage_products_ordered = (tot_products_ordered / tot_products) * 100
#create tuple
order_status = (tot_products_ordered, percentage_products_ordered)

print(type(order_status))

print(f"Order Statistics:\n"
      f"Total Products Ordered: {tot_products_ordered}\n"
      f"Percentage of Products Ordered: {percentage_products_ordered}%")
```

```python
#ordered items in customer_orders
#substract customer_order of all products


updated_inventory = {product: (quantity - 1 if product in customer_orders else quantity) for product, quantity in inventory.items()}
#print each item on a new line
print("Updated Inventory:")
for product, quantity in updated_inventory.items():
    print(f"{product}: {quantity}")
```
